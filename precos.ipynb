```python
import pandas as pd
import requests
```

```python
old = requests.get('https://www.gov.br/anp/pt-br/assuntos/precos-e-defesa-da-concorrencia/precos/precos-revenda-e-de-distribuicao-combustiveis/shlp/2001-2012/semanal-brasil-2004-a-2012.xlsx')
new = requests.get('https://www.gov.br/anp/pt-br/assuntos/precos-e-defesa-da-concorrencia/precos/precos-revenda-e-de-distribuicao-combustiveis/shlp/semanal/semanal-brasil-desde-2013.xlsx')
```

```python
# path_old = 'semanal-brasil-2004-a-2012.xlsx' 
# path_new = 'semanal-brasil-desde-2013.xlsx'

df1 = pd.read_excel(old.content, skiprows = [x for x in range(12)])
df2 = pd.read_excel(new.content, skiprows = [x for x in range(17)])
```

Fonte: https://www.gov.br/anp/pt-br/assuntos/precos-e-defesa-da-concorrencia/precos/precos-revenda-e-de-distribuicao-combustiveis/serie-historica-do-levantamento-de-precos

```python
df = pd.concat([df1, df2])
```

```python
df = df.loc[:, ~df.columns.isin(['DATA FINAL','NÚMERO DE POSTOS PESQUISADOS', 'DESVIO PADRÃO REVENDA', 'PREÇO MÍNIMO REVENDA'])]
df = df.loc[:, ~df.columns.isin(['PREÇO MÁXIMO REVENDA','MARGEM MÉDIA REVENDA', 'COEF DE VARIAÇÃO REVENDA', 'PREÇO MÉDIO DISTRIBUIÇÃO'])]
df = df.loc[:, ~df.columns.isin(['DESVIO PADRÃO DISTRIBUIÇÃO', 'PREÇO MÍNIMO DISTRIBUIÇÃO', 'PREÇO MÁXIMO DISTRIBUIÇÃO', 'COEF DE VARIAÇÃO DISTRIBUIÇÃO'])]
df
```

```python
df.rename(columns= {'DATA INICIAL':'DATA','PREÇO MÉDIO REVENDA':'PREÇO'}, inplace=True)
df
```

```python
sm = [
	['2004-05-01',260],
	['2005-05-01',300],
	['2006-04-01',350],
	['2007-04-01',380],
	['2008-03-01',415],
	['2009-02-01',465],
	['2010-01-01',510],
	['2011-01-01',545],
	['2012-01-01',622],
	['2013-01-01',678],
	['2014-01-01',724],
	['2015-01-01',788],
	['2016-01-01',880],
	['2017-01-01',937],
	['2018-01-01',954],
	['2019-01-01',998],
	['2020-01-01',1039],
	['2020-02-01',1045],
	['2021-01-01',1100],
	['2022-01-01',1212]]

dfsm = pd.DataFrame(sm, columns=['DATA','SALÁRIO MÍNIMO'])
dfsm['DATA'] = pd.to_datetime(dfsm['DATA'])
dfsm.info()
```

Fonte: https://audtecgestao.com.br/capa.asp?infoid=1336

```python
df = df.sort_values(by = 'DATA')
df = pd.merge_asof(df, dfsm, on="DATA")
df.tail(20)
```

```python
df['PODER DE COMPRA'] = df['SALÁRIO MÍNIMO'] / df['PREÇO']
df.info()
```

```python
df[df["PRODUTO"] == "GASOLINA COMUM"].plot(x="DATA", y="PODER DE COMPRA", xlabel="ANO", ylabel="LITROS DE GASOLINA / SALÁRIO MÍNIMO", legend=False, style='r', figsize=(10, 5))
df[df["PRODUTO"] == "GLP"].plot(x="DATA", y="PODER DE COMPRA", xlabel="ANO", ylabel="BOTIJÕES DE 13kg (GLP) / SALÁRIO MÍNIMO", legend=False, style='g', figsize=(10, 5))
df[df["PRODUTO"] == "GNV"].plot(x="DATA", y="PODER DE COMPRA", xlabel="ANO", ylabel="m3 de gás (GNV) / SALÁRIO MÍNIMO", legend=False, style='b', figsize=(10, 5))
df[(df["PRODUTO"] == "ÓLEO DIESEL") | (df["PRODUTO"] == "OLEO DIESEL")].plot(x="DATA", y="PODER DE COMPRA", xlabel="ANO", ylabel="LITROS DE DIESEL / SALÁRIO MÍNIMO", legend=False, style='r', figsize=(10, 5))
df[df["PRODUTO"] == "ETANOL HIDRATADO"].plot(x="DATA", y="PODER DE COMPRA", xlabel="ANO", ylabel="LITROS DE ETANOL / SALÁRIO MÍNIMO", legend=False, style='g', figsize=(10, 5))
```
